### Price and Odometer can probably lose the dollar sign and km respectively 

```python
autos['price'] = (autos['price']
                  .str.replace('$','')
                  .str.replace(',','')
                  .astype(float)
                 )
```

## Ford and Open are less Expesnive

```python
avg_mileage = {}
for b in brand_types:
    brand_only = autos[autos['brand'] == b]
    avg_mileage[b] = int(brand_only['odometer_km'].mean())
    #avg_mileage[b] = brand_only['price'].mean()
    
avg_mileage
```

```python
avg_price = {}
for b in brand_types:
    brand_only = autos[autos['brand'] == b]
    #avg_price[b] = brand_only['odometer_km'].mean()
    avg_price[b] = int(brand_only['price'].mean())
    
avg_price
```

```python
brand_avg_mileage = pd.Series(avg_mileage) 
brand_avg_price = pd.Series(avg_price)
print(brand_avg_price)
```
